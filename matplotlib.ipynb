```python
import numpy as np
x=np.arange(0,10)
y=x*x
x1 =[1,2,3,4]
y1=[3,4,5,6]
plt.bar(x,y,color='g')
plt.bar(x1,y1,color='r')
plt.show()
```

```python
import numpy as np
x=np.arange(0,5)
y=x*x
x1 =[1,2,3,4]
y1=[3,4,5,6]
plt.hist(x,y,color='g')
plt.hist(x1,y1,color='b')
plt.show()
```
